```python
#challenge7
from sklearn.learning_curve import learning_curve
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
model = LogisticRegression()

X=df[df.columns[1:]]
y=df[df.columns[0]]
title="Learning curve for Logistic regression model"

#learning_curve(model,X,y,train_sizes=[3,6,10],cv=3)
nrCrossValidationFolds = 5###k
nrSplits = 10
trainSplits = np.linspace(.05, 1.0, nrSplits)
train_sizes, train_errors, test_errors=learning_curve(model, X, y, train_sizes=trainSplits,
                                                      cv=nrCrossValidationFolds, n_jobs=1)
mean_train_errors = np.mean(train_errors, axis=1)
mean_test_errors = np.mean(test_errors, axis=1)

plt.plot(train_sizes, mean_train_errors, label="Training accuracy", color="b")
plt.plot(train_sizes, mean_test_errors, label="cross validation accuracy", color="g")
plt.legend(loc="lower right")
plt.grid()
print train_sizes
# print train_errors
# print test_errors

#plot_learning_curve(model, title, X, y, ylim=None, cv=None,
                       # n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5))
#model.fit(X_train, y_train)
#print model
#expected=y_test
#predicted=model.predict(X_test)
```

```python
##challenge 8

model = KNeighborsClassifier(n_neighbors=2)

title="Learning curve for nearest neighbour classifier model"


train_sizes, train_errors, test_errors=learning_curve(model, X, y, train_sizes=trainSplits,
                                                      cv=nrCrossValidationFolds, n_jobs=1)
mean_train_errors = np.mean(train_errors, axis=1)
mean_test_errors = np.mean(test_errors, axis=1)

plt.plot(train_sizes, mean_train_errors, label="training errors", color="b")
plt.plot(train_sizes, mean_test_errors, label="testing errors", color="g")
plt.title("Learning curve for K nearest neighbors model (K=2)")
plt.xlabel("Number of training examples")
plt.ylabel("Accuracy")
plt.legend(loc=4)
plt.grid()
print train_sizes
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

models = {'gaussianNB': GaussianNB(),
          'svc': SVC(probability=True),
          'decisiontree': DecisionTreeClassifier(),
          'randomforest': RandomForestClassifier() }
colour_dict={'gaussianNB': 'g',
             'svc': 'r',
             'decisiontree':'b',
             'randomforest':'y'}
plt.figure(1, (9,6))
for name,model in models.iteritems():
    train_sizes, train_errors, test_errors=learning_curve(model, X, y, train_sizes=trainSplits,
                                                          cv=nrCrossValidationFolds, n_jobs=1)
    mean_test_errors = np.mean(test_errors, axis=1)
    labeltemp=name
    colortemp=colour_dict[labeltemp]
    plt.plot(train_sizes, mean_test_errors, label=labeltemp, color=colortemp)

plt.title("Learning curves for various models")
plt.xlabel("Number of training samples")
plt.ylabel("Test Accuracy")
plt.grid()
plt.legend(loc=4)   
plt.show()
```

```python
#challenge 9
from sklearn.cross_validation import cross_val_score

print "Accuracy for various models with %d-fold cross validation:" % nrCrossValidationFolds
scores = ['','','','']
#print scores
for name,model in models.iteritems():
    scores = cross_val_score(model, X, y, cv=nrCrossValidationFolds, n_jobs=1)
    print "%s: %0.3f" % (name, scores.mean())
```

```python
#challenge 11
import pandas as pd
%matplotlib inline
moviesDF=pd.read_csv('BoxOfficeMojoMovies.csv', sep=',', header=0)
moviesDF.head()
moviesDF2 = moviesDF.dropna(subset=['ProductionBudget','RuntimeInMinutes','OpeningWeekendGross','OWNumberTheaters','OWAvgGrossPerTheater'])
moviesDF2.head()
countsPerRating = moviesDF2.groupby(['MPAARating'], sort=True).count()
countsPerRating.head()
plt = countsPerRating['Title'].plot(kind='bar', title="MPAARating frquencies")

print countsPerRating['Title']
```

```python
#challenge 11
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import cross_val_score
features =  ['ProductionBudget','RuntimeInMinutes','OpeningWeekendGross','OWNumberTheaters','OWAvgGrossPerTheater'] ###features
label = ['MPAARating']
labelValues = ['G', 'NC-17', 'PG', 'PG-13', 'R', 'Unrated']###classes
moviesDF3 = moviesDF2.replace(to_replace=labelValues,value=range(len(labelValues)))
X_values = moviesDF3[features]
y_values = moviesDF2[label[0]]

# print labelValues

# print X_values

# print y_values

models = {'logres' : LogisticRegression(),
              'knn' : KNeighborsClassifier(n_neighbors=3),
              'baseline': DummyClassifier(strategy='most_frequent')}
nrCrossValidationFolds = 5

print "Accuracy for various models with %d-fold cross validation:" % nrCrossValidationFolds

for eName, e in models.items():
    scores = cross_val_score(e, X_values, y=y_values, cv=nrCrossValidationFolds, n_jobs=1)
    print "%s: %0.3f" % (eName, scores.mean())
models['logres'].fit(X_values,y_values).coef_
```
